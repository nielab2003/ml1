```python
import numpy as pd
import pandas as pd
```

```python
df=pd.read_csv('student_clustering.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
X=df.iloc[:,:]
km=KMeans(n_clusters=4)
y=km.fit_predict(X)
```

```python
y
```

```python
plt.scatter(X.iloc[y== 0,0],X.iloc[y== 0,1],color='blue')
plt.scatter(X.iloc[y== 1,0],X.iloc[y== 1,1],color='red')
plt.scatter(X.iloc[y== 2,0],X.iloc[y == 2,1],color='green')
plt.scatter(X.iloc[y== 3,0],X.iloc[y == 3,1],color='yellow')
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X,y)
print("Silhouette Score for 4 clusters:", score)
```
