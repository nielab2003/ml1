```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("placement.csv")
df.head()
```

```python
plt.scatter(df['exp(in months)'], df['salary(in thousands)'])
plt.xlabel("Years Experience (in months)")
plt.ylabel('Salary in thousands') 
plt.title('Years of exp vs salary')
plt.show()
```

```python
# Extract the first column (YearsExperience) as X
X = df.iloc[:, 0:1]

# Extract the last column (Salary) as y
y =df.iloc[:,-1]
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)

# Create and train the model
lr = LinearRegression()
lr.fit(X_train, y_train)

print(X_test)
```

```python
# # Predict a single value (example: 29.78 months experience)
# pred_value = lr.predict(X_test.iloc[[0]])
# print("Predicted Salary:", pred_value)
print(y_test)
```

```python
plt.scatter(df['exp(in months)'], df['salary(in thousands)'])
plt.plot(X_test, lr.predict(X_test), color='black')
plt.xlabel('Years of Experience (in months)')
plt.ylabel('Salary (in thousands)')
plt.title('Years of Experience vs Salary')
plt.show()
```

```python
m = lr.coef_
print(m)
```

```python
b=lr.intercept_
print(b)
```

```python
from sklearn.metrics import mean_squared_error
y=lr.predict(X_test)
print(mean_squared_error(y_test,y))
```
