```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
data=df.iloc[:,3:5].values
data
```

```python
#for Ward's- method linkage
import scipy.cluster.hierarchy as shc
# plt.figure(figsize=(5, 5))
plt.title("Customers Dendrogram-Ward's Linkage")
graph=shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
#for ward's linkage
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5,linkage='ward')
labels_= cluster.fit_predict(data)
```

```python
labels_
```

```python
# plt.figure(figsize=(5,5))
plt.scatter(data[:,0],data[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(data, labels_)
print(f'Silhouette Score wards method: {score}')
```
